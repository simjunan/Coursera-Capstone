```python
df = pd.read_csv('C:\\Users\\User\\Desktop\\aed-locations.csv')
df
```

```python
#merge same postal code
#df = df.groupby('postal_code').agg({'operating_hours':', '.join,'house_number':', '.join,'road_name':', '.join,
 #                                  'building_name':'first','unit_number':'first','aed_location_description':', '.join, 'aed_location_floor_level':'first'}).reset_index
df = df.groupby(['postal_code']).agg(lambda x: ', '.join(x)) 
df = df.reset_index()
df
```

```python
import onemapsg
from onemapsg import OneMapClient
df['postal_code'] = df['postal_code'].astype('int64')
Client = OneMapClient("***@gmail.com", "****")
results = []
df2 = df.postal_code
for code in df2:
    results.append(Client.search(code))
results
```

```python
ref_results =[]
for list_item in results:
    if list_item['results'] != []:
        ref_results.append(list_item['results'])
ref_results
```

```python
Lat = []
Long = []
for list_element in ref_results:
    for i in list_element:
        Lat.append(i['LATITUDE'])
        Long.append(i['LONGITUDE'])
d = {'Latitude':Lat,'Longitude':Long}
df = pd.DataFrame(d)
df
```

```python
#Create map
import folium
Singapore = folium.Map(location=[df.Latitude[1], df.Longitude[1]],zoomstart=20)

#Create markers
for lat, lng in zip(df['Latitude'], df['Longitude']):
    label = '{}, {}'.format(lat, lng)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(Singapore)
Singapore
```
